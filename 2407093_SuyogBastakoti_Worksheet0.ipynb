**Create a Python program that converts between different units of measurement.**

**• The program should:**
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.
• Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.
• Conversion Options:
1. Length:– Convert meters (m) to feet (ft).– Convert feet (ft) to meters (m).
2. Weight:– Convert kilograms (kg) to pounds (lbs).– Convert pounds (lbs) to kilograms (kg).
3. Volume:– Convert liters (L) to gallons (gal).– Convert gallons (gal) to liters (L)

```python
def convert_units(conversion_type, value, direction):
    """
    Converts values between different units of measurement.

    Parameters:
    conversion_type (str): The type of conversion ('length', 'weight', 'volume').
    value (float): The numeric value to convert.
    direction (str): The conversion direction:
                     For length: 'm_to_ft' or 'ft_to_m'
                     For weight: 'kg_to_lbs' or 'lbs_to_kg'
                     For volume: 'l_to_gal' or 'gal_to_l'

    Returns:
    float: The converted value.
    """
    # Conversion factors
    if conversion_type == "length":
        if direction == "m_to_ft":
            return value * 3.28084
        elif direction == "ft_to_m":
            return value / 3.28084
    elif conversion_type == "weight":
        if direction == "kg_to_lbs":
            return value * 2.20462
        elif direction == "lbs_to_kg":
            return value / 2.20462
    elif conversion_type == "volume":
        if direction == "l_to_gal":
            return value * 0.264172
        elif direction == "gal_to_l":
            return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type or direction.")

# Main program
print("Welcome to the Unit Converter!")

try:
    conversion_type = input("Choose conversion type (length, weight, volume): ").strip().lower()

    if conversion_type not in ["length", "weight", "volume"]:
        raise ValueError("Invalid conversion type selected.")

    value = float(input("Enter the value to convert: "))

    if conversion_type == "length":
        direction = input("Convert 'm_to_ft' or 'ft_to_m'?: ").strip().lower()
    elif conversion_type == "weight":
        direction = input("Convert 'kg_to_lbs' or 'lbs_to_kg'?: ").strip().lower()
    elif conversion_type == "volume":
        direction = input("Convert 'l_to_gal' or 'gal_to_l'?: ").strip().lower()

    result = convert_units(conversion_type, value, direction)
    print(f"Converted value: {result:.2f}")

except ValueError as ve:
    print(f"Error: {ve}")
except Exception as e:
    print(f"Unexpected error: {e}")
```

```python
# 1. Flatten a Nested List
def flatten(lst):
    """
    Flattens a nested list into a single list (one level).
    """
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result


# 2. Access Element from Nested List
def access_nested_element(lst, indices):
    """
    Access an element from a nested list using a list of indices.
    """
    element = lst
    for i in indices:
        element = element[i]
    return element


# 3. Sum of All Elements in Nested List (any depth)
def sum_nested(lst):
    """
    Returns the sum of all numbers in a nested list regardless of depth.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# 4. Remove Specific Element from Nested List
def remove_element(lst, elem):
    """
    Removes all occurrences of elem from a nested list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result


# 5. Find Maximum Element in Nested List
def find_max(lst):
    """
    Finds the maximum element in a nested list regardless of depth.
    """
    max_val = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)
    return max_val


# 6. Count Occurrences in Nested List
def count_occurrences(lst, elem):
    """
    Counts how many times elem appears in a nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


# 7. Deep Flatten (any depth)
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result


# 8. Average of Nested List
def average_nested(lst):
    """
    Returns the average of all numbers in a nested list.
    """
    flat_list = deep_flatten(lst)
    return sum(flat_list) / len(flat_list)


# ---------------------------
# Example Runs
# ---------------------------

print(flatten([[1,2],[3,4],[5]]))

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))

print(sum_nested([[1,2],[3,[4,5]],6]))

print(remove_element([[1,2],[3,2],[4,5]],2))

print(find_max([[1,2],[3,[4,5]],6]))

print(count_occurrences([[1,2],[2,3],[2,4]],2))

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))

print(average_nested([[1,2],[3,4],[5,6]]))
```

```python
#First import NumPy and define the matrices.
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)
```

```python
#1. Prove 𝐴⋅𝐴−1=𝐼A⋅A−1=I
# Find inverse of A
A_inv = np.linalg.inv(A)

print("Inverse of A:")
print(A_inv)

# Multiply A and A inverse
result = np.dot(A, A_inv)

print("\nA * A^-1 :")
print(np.round(result,5))
```

**The result shows that multiplying a matrix by its inverse gives the identity matrix.**

```python
#2. Prove 𝐴𝐵≠𝐵A
AB = np.dot(A,B)
BA = np.dot(B,A)

print("\nAB =")
print(AB)

print("\nBA =")
print(BA)
```

```python
#3. Prove (𝐴𝐵)𝑇=BTAT
# Left side
left = np.transpose(np.dot(A,B))

# Right side
right = np.dot(np.transpose(B), np.transpose(A))

print("\n(AB)^T =")
print(left)

print("\nB^T A^T =")
print(right)
```

Both outputs will be the same, proving:

**(𝐴𝐵)𝑇=BTAT**

```python
#Given equations:

# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Matrix form:

# 𝐴𝑋 = B

# Where:

# A = coefficient matrix
# X = [x y z]
# B = constant matrix

#Step 1: Create matrices
A = np.array([
    [2,-3,1],
    [1,-1,2],
    [3,1,-1]
])

B = np.array([-1,-3,9])
```

```python
#importing neccessary libraries
import numpy as np
import time
```

```python
# 1. Element-wise Addition
size = 1_000_000

# Python lists
list1 = list(range(size))
list2 = list(range(size))

start = time.time()
list_add = [a + b for a, b in zip(list1, list2)]
end = time.time()
print("Python List Addition Time:", round(end-start, 5), "seconds")

# NumPy arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start = time.time()
np_add = arr1 + arr2
end = time.time()
print("NumPy Array Addition Time:", round(end-start, 5), "seconds")
```

```python
# 2. Element-wise Multiplication
# ---------------------------

# Python lists
start = time.time()
list_mul = [a * b for a, b in zip(list1, list2)]
end = time.time()
print("\nPython List Multiplication Time:", round(end-start, 5), "seconds")

# NumPy arrays
start = time.time()
np_mul = arr1 * arr2
end = time.time()
print("NumPy Array Multiplication Time:", round(end-start, 5), "seconds")
```

**Summary:**

**From this experiment, we can see a huge difference in performance between plain Python lists and NumPy arrays:**

Element-wise operations (addition and multiplication) are much faster with NumPy, even for large arrays of 1,000,000 elements.

Python lists use slow Python loops, while NumPy uses optimized C/Fortran routines internally.

Dot product calculations also show that NumPy is orders of magnitude faster than plain Python lists.

Matrix multiplication highlights the biggest difference:

Python lists take several minutes for large matrices, and sometimes can even freeze the notebook.

NumPy performs the same operation in fractions of a second, even for 1000×1000 matrices.

Reason: NumPy arrays are contiguous blocks of memory, support vectorized operations, and use highly optimized linear algebra libraries like BLAS and LAPACK. Python lists, on the other hand, are high-level objects and not optimized for numerical computations.

**Overall:**

This experiment clearly demonstrates that for numerical computations, NumPy is far more efficient and practical than Python lists, making it the standard choice for data analysis, scientific computing, and machine learning tasks.
